## Resolution

There are four types of resolution to consider for any dataset:

1. Radiometric
2. Spectral
3. Spatial
4. Temporal


 

### Radiometric Resolution

- Radiometric resolution refers to the amount of information in each pixel. i.e. the number of bits representing the energy recorded
- The higher the radiometric resolution, the more the values available to store information, providing better discrimination between even the slightest differences in energy




### Spatial Resolution

- Defined by the size of each pixel within a digital image, and the area on Earth's surface represented by that pixel
- The finer the resolution (the lower the number), the more detail you can see



### Spectral Resolution

- Is the ability of an instrument to discern finer wavelengths: Having more and narrower bands
- Many instruments are considered multispectral i.e. have 3 to 10 bands
- Some instruments have hundreds or thousands of bands and are considered hyperspectral
- The narrower the range of wavelengths for a given band, the finer the resolution
- The Airborne Visible/Infrared Imaging Spectrometer (AVIRIS) captures information in 224 spectral channels. At this level of detail, distinctions can be made between vegetation types, rock and mineral types, etc



### Temporal Resolution

- Temporal resolution is the time it takes a space-based platform to complete an orbit and revisit the same observation area
- Temporal resolution depends on the orbit, the instrument's characteristics, and the swath width

It is difficult to combine all of the desirable features into one remote instrument. Researchers have to make trade-offs
- When researching weather, which is dynamic over time, a high temporal resolution is critical
- When researching seasonal vegetation changes, on the other hand, a high temporal resolution may be sacrificed for a higher spectral or spatial resolution
